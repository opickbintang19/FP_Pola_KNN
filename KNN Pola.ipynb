# Dataset

```python
import pandas as pd
dataset = pd.read_csv('dataset.csv')
dataset
```

# Data CSV --> Array

```python
import numpy as np

dataset.to_numpy()
```

# Input Data

Keterangan :

x1 = Mendengarkan (400)

x2 = Merespon Kaidah (450)

x3 = Membaca (304)

x4 = Menulis (4)

x5 = Berbicara (4)

K = (3)

```python
x1 = int(input("Nilai Mendengarkan :"))
x2 = int(input("Nilai Merespon Kaidah :"))
x3 = int(input("Nilai Membaca :"))
x4 = int(input("Nilai Menulis :"))
x5 = int(input("Nilai Berbicara :"))
KReturn = int(input("K : "))

X_new = np.array([x1,x2,x3,x4,x5]).reshape(1,-1)
X_new
```

# Klasifikasi dengan KNN

```python
X_train = np.array(dataset[['Mendengarkan','Merespon Kaidah','Membaca','Menulis','Berbicara']])
y_train = np.array(dataset['Kesimpulan'])

print(f'X_train:\n{X_train}\n')
print(f'y_train: {y_train}')
```

# Training KNN

```python
from sklearn.neighbors import KNeighborsClassifier

K = KReturn
model = KNeighborsClassifier(n_neighbors=K)
model.fit(X_train,y_train)
```

# Kalkulasi Euclidean Distance

```python
from scipy.spatial.distance import euclidean

data_jarak = [euclidean(X_new,d) for d in X_train]
dataset['Jarak'] = data_jarak
dataset.sort_values(['Jarak'])
```

```python
y_new = model.predict(X_new)
y_new
```
